```python
import seaborn as sns
%matplotlib inline
```

```python
plt.scatter(x ='age',y='restecg',data=ds,c='g')
plt.scatter(x='age',y='chol',data=ds,c='r',marker='+')
plt.ylabel('age')
plt.ylabel('chol and restecg')
plt.legend(title ='scatter_plot')
plt.show()
```

```python
plt.figure()
sns.lmplot(x ='age',y='chol',data=ds,fit_reg =False)
```

```python
plt.figure()
ds.plot(kind ='hist')
```

```python
plt.figure()
sns.displot(ds.age)
```

```python
plt.figure()
plt.xlabel('cp')
plt.ylabel('count')
ds.cp.value_counts().plot(kind='bar')
```

```python
plt.figure()
sns.countplot(ds.cp)
```

```python
plt.figure()
sns.lineplot(x='age',y='chol',data=ds)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize =(8,6))
ds['cp'].value_counts().plot(kind='pie',autopct='%1.1f%%',startangle =10)
plt.title('Distribution of chest Pain Types')
plt.show()
```
