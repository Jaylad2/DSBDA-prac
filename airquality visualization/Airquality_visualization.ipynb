```python
import seaborn as sns
%matplotlib inline #tells Jupyter to render the plots directly within
#the notebook's output cell.
```

```python
plt.scatter(x='Ozone',y='Solar.R',data=ds,c='g')
plt.scatter(x='Ozone',y='Solar.R',data=ds,c='r',marker='+')
plt.xlabel('Ozone')
plt.ylabel('Ozone and Solar.R')
plt.legend(title='scatter_plot')
plt.show()
```

```python
plt.figure()
sns.lmplot(x='Ozone',y='Solar.R',data=ds,fit_reg=False)
plt.show()
```

```python
plt.figure()
ds.plot(kind='hist')
```

```python
plt.figure()
sns.displot(ds.Wind)
```

```python
plt.figure()
plt.xlabel('Month')
plt.ylabel('count')
ds.Month.value_counts().plot(kind='bar')
```

```python
plt.figure()
sns.countplot(ds.Month)
```

```python
plt.figure()
sns.lineplot(x='Temp',y='Month',data =ds)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
ds['Month'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Distribution of Months')
plt.show()
```
